## feature scaling
primarly ther are two type
1.min-maax scaling(normalization)
(value-min)/(max-min)=iska value hameza 0-1 ke beech me lyi karega
class is minmaxscaler
2.standardization
(value-min)/standarddaviation
sklearn provide class called standardscaler



pipeline ka matlab series of step krenge

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline=Pipeline([
      ('imputer',SimpleImputer(strategy='median')),
      ('std_scaler',StandardScaler()),
])
```

```python
housing_num_tr=my_pipeline.fit_transform(housing)
```

```python
housing_num_tr
```

## selecting a desired model for dragon real estates

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model=DecisionTreeRegressor()
model=RandomForestRegressor()
model.fit(housing_num_tr,housing_labels)
```

 ## evaluating model

```python
from sklearn.metrics import mean_squared_error
import numpy as np
housing_predictions=model.predict(housing_num_tr)
mse=mean_squared_error(housing_labels,housing_predictions)
rmse=np.sqrt(mse)
```

## using better evaluation technique-cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(model,housing_num_tr,housing_labels,scoring="neg_mean_squared_error",cv=10)
```

```python
rmse_score=np.sqrt(-score)
```

```python
rmse_score
```

## making function

```python
def print_score(score):
    print("score:",score)
    print("mean:",score.mean())
    print("standard deviation:",score.std())
```

```python
print_score(rmse_score)
```
